In this file, collect the number of academy awards that actor or director acheived  

```python
from IPython.display import HTML
import base64
import pandas as pd

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

from bs4 import BeautifulSoup
```

```python
names = []
years = []
roles = []
```

```python
options = Options()
options.add_argument("start-maximized")

webdriver_service = Service('E:\chromedriver.exe')
driver = webdriver.Chrome(options=options, service=webdriver_service)
```

```python
for i in range(1934, 2023):
    print(i)
    url = f'https://www.imdb.com/event/ev0000003/{i}/1/?ref_=ev_eh'
    driver.get(url)
    delay = 30 # seconds
    try:
        myElem = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.ID, 'pagecontent')))
        print("Page is ready!")
    except TimeoutException:
        print("Loading took too much time!")
    the_soup = BeautifulSoup(driver.page_source, 'html.parser')
    widgets = the_soup.find_all('div', class_='event-widgets__award')
    widget = widgets[0]
    categories = widget.find_all('div', class_='event-widgets__award-category')
    for category in categories:
      name = str(category.find('div', class_='event-widgets__award-category-name').text).lower()
      print(name)
      if 'actor' in name or 'actress' in name or 'director' in name or 'directing' in name:
        events = category.find_all('div', class_='event-widgets__award-nomination')
        win = events[0]
        names.append(win.find('span', class_='event-widgets__nominee-name').text)
        years.append(i)
        if 'actor' in name:
          roles.append('actor')
        elif 'actress' in name:
          roles.append('actress')
        elif 'director' in name or 'directing' in name:
          roles.append('director')
      else:
        continue
```

```python
oscar = pd.DataFrame({'year': years, 'name': names, 'role': roles})
```

```python
oscar.head()
```

```python
# download file 
title = "Download CSV file"
filename = "oscar_award.csv"
csv = oscar.to_csv(index=False)
b64 = base64.b64encode(csv.encode())
payload = b64.decode()
html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
html = html.format(payload=payload,title=title,filename=filename)
HTML(html)
```
